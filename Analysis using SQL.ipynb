```python
%%sql
SELECT * FROM facts 
order by birth_rate DESC
limit 5;
```

```python
%%sql
SELECT
    MIN(population) min_pop,
    MAX(population) max_pop, 
    MIN(population_growth) min_pop_grwth,
    MAX(population_growth) max_pop_grwth 
FROM facts;
```

```python
%%sql
select name from facts
order by population 
limit 5
```

```python
%%sql
select * from facts
where population = (select min(population) from facts)
```

```python
%%sql
select * from facts
where population = (select max(population) from facts)
```

```python
%%sql
SELECT
    AVG(population) avg_population,
    AVG(area) avg_area
FROM facts;
```

```python
%%sql
select name from facts
where population > (SELECT AVG(population) from facts)
```

```python
%%sql
SELECT * FROM facts
WHERE population > (
    SELECT AVG(population)
    FROM facts)
  AND 
    area < (
    SELECT AVG(area)
    FROM facts)
```
